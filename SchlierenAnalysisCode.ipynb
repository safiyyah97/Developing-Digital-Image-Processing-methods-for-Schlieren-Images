<h2> 1. Raw image to Sobel image <h2>

```python
## Importing the libraries 
import cv2
import fnmatch 
import os 

## Change parameters  
dir_path = ' '                                         # Working directory file path
kernel_size = 7                                        # Chosen kernel size
binary_threshold = 190                                 # Chosen binary thresold

reference_picture = '/Ref.png'                         # Name of reference image
scale_picture = '/Scale.png'                           # Name of scale image 

picture_name = '/I ('                                  # Name of image 
format_name = ').png'                                  # Format

crop_folder = '/crop'                                  # Crop image folder name 
newpath1 = str(dir_path) + str(crop_folder)  
picture_name_1 = '/Icrop ('                            # Name of cropped image

div_folder = '/div'                                    # Divided image folder name
newpath2 = str(dir_path) + str(div_folder)    
picture_name_2 = '/Idiv ('                             # Name of divided image

sobelx_folder = '/sobelx'+ str(kernel_size)            # Sobel image folder name 
newpath3 = str(dir_path) + str(sobelx_folder)    
picture_name_3 = '/Isobelx ('                          # Name of sobelx image

binary_folder = '/binary'                              # Binary image folder name
newpath4 = str(dir_path) + str(binary_folder)    
picture_name_4 = '/Ibinary ('                          # Name of binary image 


M = len(fnmatch.filter(os.listdir(dir_path), '*).png*'))
Iref = cv2.imread(str(dir_path) + str(reference_picture))

cv2.namedWindow('Reference image',2)
cv2.resizeWindow('Reference image', 1024, 544)
    #cv2.namedWindow("Image",2)
ROI = cv2.selectROI('Reference image', Iref, False, False)
Iref_crop = Iref[int(ROI[1]):int(ROI[1]+ROI[3]), int(ROI[0]):int(ROI[0]+ROI[2])]
cv2.imwrite(str(dir_path)+ str(crop_folder) + '/Refcrop.png', Iref_crop)
cv2.waitKey()
cv2.destroyAllWindows()

k = 220                                                    # Image number 

# CROP IMAGE 
picture = (str(dir_path)+str(picture_name)+str(k)+str(format_name))
I = cv2.imread(picture)
Icrop = I[int(ROI[1]):int(ROI[1]+ROI[3]), int(ROI[0]):int(ROI[0]+ROI[2])]    
I = cv2.rotate(Icrop, cv2.ROTATE_90_COUNTERCLOCKWISE)
Iref = cv2.rotate(Iref_crop, cv2.ROTATE_90_COUNTERCLOCKWISE)
if not os.path.exists(newpath1):
    os.makedirs(newpath1)
    cv2.imwrite(str(newpath1)+str(picture_name_1)+str(k)+str(format_name), I)
else: 
    cv2.imwrite(str(newpath1)+str(picture_name_1)+str(k)+str(format_name), I)

# NORMALISED IMAGE    
Idiv = (I/Iref)*255
if not os.path.exists(newpath2):
    os.makedirs(newpath2)
    cv2.imwrite(str(newpath2)+str(picture_name_2)+str(k)+str(format_name), Idiv)
else: 
    cv2.imwrite(str(newpath2)+str(picture_name_2)+str(k)+str(format_name), Idiv)

# BINARASED IMAGE    
Ibinary = cv2.threshold(Idiv, binary_threshold, 250, cv2.THRESH_BINARY_INV)[1]  # ensure binary
if not os.path.exists(newpath4):
    os.makedirs(newpath4)
    cv2.imwrite(str(newpath4)+str(picture_name_4)+str(k)+str(format_name), Ibinary)
else: 
    cv2.imwrite(str(newpath4)+str(picture_name_4)+str(k)+str(format_name), Ibinary)

# SOBEL IMAGE     
sobelx = cv2.Sobel(Ibinary,cv2.CV_64F,1,0,ksize=kernel_size)  
if not os.path.exists(newpath3):
    os.makedirs(newpath3)
    cv2.imwrite(str(newpath3)+str(picture_name_3)+str(k)+str(format_name), sobelx)
else:
    cv2.imwrite(str(newpath3)+str(picture_name_3)+str(k)+str(format_name), sobelx)  

print('Finished')
```

<h2> 5. Add scale bar and timestamp (this is for a series of images) <h2>

```python
import cv2
from datetime import datetime
import os
# File Path
dir_path = 'E:/Belgium (03.2022)/Processed Schlieren data/13.05.22'

# New folder
scale_time_folder = '/scale_time'

# Timestamp textfile 
text_file = 'timestamps.txt'

# Scale: Number of pixels in 1 cm
pixel_scale = 110 

def add_timestamp(input_picture, output_picture, timestamp):
    # Read the image
    img = cv2.imread(input_picture)
    
    #This is for an image that is rotated 
    img = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
    img_height = img.shape[0]
    img_width = img.shape[1]
  
    # Define the font and other text properties
    font = cv2.FONT_HERSHEY_SIMPLEX
    font_scale = 2
    font_thickness = 3
    text_color = (255, 255, 255)  # White color in BGR

    # Define the position to add the timestamp
    position = (int(img_width)-300, int(img_height)-30)
    # Add timestamp to the image
    cv2.putText(img, str(time_format), position, font, font_scale, text_color, font_thickness, cv2.LINE_AA)

    row, col = img.shape[:2]
    bottom = img[row-2:row, 0:col]
    mean = cv2.mean(bottom)[0]
    border_size = 20
    border = cv2.copyMakeBorder(img, top=border_size, bottom=border_size, left=border_size, right=border_size, borderType=cv2.BORDER_CONSTANT, value=[mean, mean, mean])
    
    img = border
    border_color = (255, 255, 255)  # White color
    border_thickness = 20
    dash_length = pixel_scale  # Length of dashes
    gap_length = pixel_scale   # Length of gaps between dashes
    for x in range(0, img.shape[1], dash_length + gap_length):
        cv2.line(img, (x, 0), (x + dash_length, 0), border_color, border_thickness)
        cv2.line(img, (x, img.shape[0] - 1), (x + dash_length, img.shape[0] - 1), border_color, border_thickness)

# Draw dashed lines at the left and right sides
    for y in range(0, img.shape[0], dash_length + gap_length):
        cv2.line(img, (0, y), (0, y + dash_length), border_color, border_thickness)
        cv2.line(img, (img.shape[1] - 1, y), (img.shape[1] - 1, y + dash_length), border_color, border_thickness)

# # Display the result
#     cv2.imshow('Dashed Border', image)
#     cv2.waitKey(0)
#     cv2.destroyAllWindows()
    cv2.imwrite(output_picture, img)

# Read timestamps from the text file
with open(str(dir_path) + '/' + str(text_file), 'r') as file:
    timestamps = file.readlines()

# Process each image and add corresponding timestamp
for i, timestamp in enumerate(timestamps):
    timestamp = int(float(timestamp.strip()))  # Remove leading/trailing whitespace
    timestamp_total_seconds = int(timestamp/1000)
    timestamp_hours = int(timestamp_total_seconds/3600)
    timestamp_minutes = int((timestamp_total_seconds - timestamp_hours*3600)/60)
    timestamp_seconds = int((timestamp_total_seconds - timestamp_hours*3600 - timestamp_minutes*60))
    
    time_format = "{:02}:{:02}:{:02}".format(timestamp_hours, timestamp_minutes, timestamp_seconds)
    
    print(time_format)
    input_picture = (str(dir_path) + '/I ('+ str(i+1) + ').png')

    if not os.path.exists(str(dir_path) + str(scale_time_folder)):
        os.makedirs(str(dir_path) + str(scale_time_folder))
        output_picture = (str(dir_path) + str(scale_time_folder) + '/Itime ('+ str(i+1) + ').png')
        
    else: 
        output_picture = (str(dir_path) + str(scale_time_folder) + '/Itime ('+ str(i+1) + ').png')


    #output_picture = (str(dir_path) + '/timestamp' + '/Itime ('+ str(i+1) + ').png')
    add_timestamp(input_picture, output_picture, timestamp)

print("Timestamps added to images.")
```
